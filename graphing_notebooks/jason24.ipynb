```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
movie = "all_movie_data.csv"
movie_data_df = pd.read_csv(movie)
movie_data_df.head()
```

```python
tweet_count = movie_data_df["tweet_count"] > 10
```

```python
movie_data_df[tweet_count].head()
```

```python
fig = plt.figure(dpi=150)
plt.ylim(-1, 1)
plt.xlim(0, 310000000)

plt.scatter(movie_data_df["budget"], movie_data_df["compound"] ,linewidths = 1, alpha= .25, color = 'dodgerblue',edgecolor='black',s=90)
plt.title("Twitter Vader Sentiment vs Movie Total Budget")

ax=plt.gca()
ax.annotate('TheEmojiMovie', xy=(50000000, 0.037363), xytext=(-0.2, -1),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel("Movie Budget in $100's Millions")
plt.ylabel("Twitter Compound Score")

plt.show()
```
