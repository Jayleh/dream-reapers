```python
# Dependencies
import glob
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set folder path
folder_path = "daily_movie_sentiment"

# Set folder
all_files = glob.glob(os.path.join(folder_path, "*.csv"))
all_files
```

```python
# Concat all csv files in folder
daily_sent_df = pd.concat((pd.read_csv(f) for f in all_files))

# Reset index
daily_sent_df = daily_sent_df.reset_index(drop=True)
daily_sent_df.head()
```

```python
# Rename columns
daily_sent_df = daily_sent_df.rename(columns={"Search Date": "Search Until Date"})

# Format date into datetime
# daily_sent_df["Search Until Date"] = pd.to_datetime(daily_sent_df["Search Until Date"], format="%Y-%m-%d")
daily_sent_df.dtypes
```

```python
# Sort by title
sorted_df = daily_sent_df.sort_values(["Movie Title", "Search Until Date"])

# Reset index
sorted_df = sorted_df.reset_index(drop=True)

# Save and show
sorted_df.to_csv("daily_movie_sentiment.csv", encoding="utf-8", index=False)
sorted_df
```

```python
# Grab movie list
movie_list = sorted_df['Movie Title'].unique()
movie_list
```

```python
# Set movie title to index
title_index = sorted_df.set_index(['Movie Title'])

# Format dates back to strings
# title_index['Search Until Date'] = title_index['Search Until Date'].astype(str)

# Subset movie dataframe into dfs for each movie
acrimony_df = title_index.loc['#Acrimony',['Compound', 'Search Until Date', 'Tweet Count']]
gringo_df = title_index.loc['#Gringo',['Compound', 'Search Until Date', 'Tweet Count']]
simon_df = title_index.loc['#LoveSimon',['Compound', 'Search Until Date', 'Tweet Count']]
pacific_df = title_index.loc['#PacificRimUprising',['Compound', 'Search Until Date', 'Tweet Count']]
prey_df = title_index.loc['#PreyAtNight',['Compound', 'Search Until Date', 'Tweet Count']]
player_df = title_index.loc['#ReadyPlayerOne',['Compound', 'Search Until Date', 'Tweet Count']]
gnomes_df = title_index.loc['#SherlockGnomes',['Compound', 'Search Until Date', 'Tweet Count']]
hurricane_df = title_index.loc['#TheHurricaneHeist',['Compound', 'Search Until Date', 'Tweet Count']]
raider_df = title_index.loc['#TombRaider',['Compound', 'Search Until Date', 'Tweet Count']]
wrinkle_df = title_index.loc['#WrinkleInTime',['Compound', 'Search Until Date', 'Tweet Count']]
```

# Movies Released on March 9th, 2018

```python
# Set figure size
plt.figure(figsize=(15, 10))

# Create line plots
gringo, = plt.plot(gringo_df['Search Until Date'], gringo_df['Compound'], label='#Gringo', ls='--', lw=0.8)
prey, = plt.plot(prey_df['Search Until Date'], prey_df['Compound'], label='#PreyAtNight', ls='--', lw=0.8)
hurricane, = plt.plot(hurricane_df['Search Until Date'], hurricane_df['Compound'], label='#TheHurricaneHeist', ls='--', lw=0.8)
wrinkle, = plt.plot(wrinkle_df['Search Until Date'], wrinkle_df['Compound'], label='#WrinkleInTime', ls='--', lw=0.8)

# Create scatter tweet size
plt.scatter(gringo_df['Search Until Date'], gringo_df['Compound'], s=gringo_df['Tweet Count'], alpha=0.8)
plt.scatter(prey_df['Search Until Date'], prey_df['Compound'], s=prey_df['Tweet Count'], alpha=0.8)
plt.scatter(hurricane_df['Search Until Date'], hurricane_df['Compound'], s=hurricane_df['Tweet Count'], alpha=0.8)
plt.scatter(wrinkle_df['Search Until Date'], wrinkle_df['Compound'], s=wrinkle_df['Tweet Count'], alpha=0.8)

# Figure labels
plt.title('Daily Twitter Sentiment (Released on March 9th, 2018)', fontsize=20)
# plt.xlabel('Date', fontsize=18)
plt.ylabel('Tweet Polarity', fontsize=18)

# Legend
plt.legend(handles=[gringo, prey, hurricane, wrinkle], fontsize=14, loc='best')

# Set axes background color
ax = plt.gca()
ax.set_facecolor('whitesmoke')

# Insert grid lines and set behind plot elements
ax.grid(color='white')
ax.set_axisbelow(True)

# Save and show
plt.savefig('../images/PredictionDailySentiment_20180309.png')
plt.show()
```
